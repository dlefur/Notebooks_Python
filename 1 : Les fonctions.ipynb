## Définition

```python
nom="g"

def f(x) :  # ne pas changer f sur cette ligne
    return 2*x**2-5*x+1

print(f"Exemple, pour x=0, on a {nom}(0) = {f(0)}.")
```

## Tableau de valeurs

```python
xmin = -4
xmax = 5
pas = 1

x=xmin
while x <=xmax :
    print(f"{nom}({x:5.1f}) = {f(x):8.2f}")
    x=x+pas
```
